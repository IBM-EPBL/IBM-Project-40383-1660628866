```python
sb.boxplot(myData['CreditScore'],data=myData)
```

```python
myData['CreditScore'].hist()
```

```python
Q1 = myData['CreditScore'].quantile(0.25)
Q3 = myData['CreditScore'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
myData['CreditScore']=np.where(myData['CreditScore']>upper_whisker,upper_whisker,np.where(myData['CreditScore']<lower_whisker,lower_whisker,myData['CreditScore']))
```
